country_region = df_carbon_tax.groupby(['Jurisdiction covered','World Bank region'], as_index=False).sum()
reg_mean = country_region.groupby('World Bank region',as_index=False ).mean()

```python
country_region
```

```python
(country_region['Price_Per_Ton_USD']==0) & (country_region['World Bank region']== 'East Asia & Pacific ') 
```

```python
avg_reg_price = {}

for region in list(reg_mean['World Bank region']):
    
    avg_reg_price[region] =  float(reg_mean['Price_Per_Ton_USD'][reg_mean['World Bank region'] == region])

avg_reg_price['Europe and Central Asia']
```

```python
for region in avg_reg_price:
    
    country_region['Price_Per_Ton_USD'+region] = np.where((country_region['Price_Per_Ton_USD']==0) \
                                                   & (country_region['World Bank region']== region),\
                                                 avg_reg_price[region], 0)
    
country_region
    
```

```python
price_columns = ['Price_Per_Ton_USDEast Asia & Pacific ',
       'Price_Per_Ton_USDEurope and Central Asia',
       'Price_Per_Ton_USDLatin America and the Caribbean  ',
       'Price_Per_Ton_USDNorth America',
       'Price_Per_Ton_USDSub-Saharan Africa']

for col in price_columns:
    
    
    country_region['Price_Per_Ton_USD'] += country_region[col]
    

np.sum(country_region['Price_Per_Ton_USD'].isna())

```
